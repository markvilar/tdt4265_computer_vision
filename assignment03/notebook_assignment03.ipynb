### e - Methods for improving performance

#### Filter sizes
Smaller filter sizes should in theory be better, since the network can do the same amount of filtering with a smaller number of parameters. This is done by having smaller filters, but more of them. However, when experimenting with the network architectures there was not a significant difference between the networks with 3x3 filters, 5x5 filters and 7x7 filters. This might be due to the model complex, i.e. the number of parameters, not being kept between architectures. If the number of filters were calculated so that the model complexity would have stayed the same, the difference might have been clearer.

#### Number of filters
The number of filters in the network did seem to have quite a bit of difference to say. With the same filter sizes, but only having 32 filters in each the convolutional layer instead of 128 (and keeping all other parameters the same), the accuracy of the network was decreased by around 5-6 %. Doing the same but decreasing the number of filters to 64 in each convolutional layer yielded a decrease in accuracy of about 2 %. This is most likely due to the fact that the number of filters is the main contributor to the number of parameters in the feature extractor and thus to model complexity.

#### Pooling vs strided convolutions
Pooling vs strided convolutions did affect both the training and performance of the network quite a lot. Generally networks with maxpooling trained better and reached a higher final test accuracy. A reason for this might be that the maxpool layers have no learnable parameters, and hence they can reduce the dimensionality of the features without adding model complexity.

#### Batch normalization
Batch normalization improved the training rate of the CNN a lot, especially for deeper architectures. A reason for might that batch normalization makes the input to the convolutional layers smaller and that makes the changes in the network parameters smaller, making the training more stable.

#### Network architecture
The first type of architecture that I tried was:

(Conv > ReLU > Batch normalization > MaxPool)xN + (Dense > Batch normalization > ReLU)xM

This architecture did not perform that well.  

The next type of architecture that I tried was:

(Batch normalization > ReLU > Conv blocks > MaxPool)xN + (Dense > Batch normalization > ReLU)xM

This architecture performed better.

#### Regularization
L2 regularization didn't have any significant effect on the network. A reason for this might be that L2 regularization mainly prevents overfitting and that this was in general not an issue in this case.

#### Weight initialization
I found weight initialization to have a significant effect on the training rate of the CNN. With Xavier initialization the CNN was consistently achieving a validation accuracy of 50 % after 1 epoch, while without it the CNN could use as much as up to 4 epochs just to get to 30 % validation accuracy. This is probably due to the weights being smaller with Xavier initialization, so that the weight changes during backpropagation become more significant in relation to the weights themselves.

#### Optimizers
The type of optimizer, along with weight initialization, was probably the most critical method for improving performance, both in terms of accuracy and training rate. By using the Adam optimizer the CNN consistenly achieved a training accuracy of 60 % after the first training epoch and generally the final accuracy was improved by around 5-10 % from the same architecture with the SGD optimizer. Part of the reason is probably that the Adam optimizer by default has momentum implemented, while the SGD optimizer does not have momentum by default. Since I used the optimizers with default momentum settings the Adam optimizer would be able to avoid local minimas, while the SGD optimizer would get stuck in them. This would help the Adam optimizer to find parameters that would decrease the loss even further.
